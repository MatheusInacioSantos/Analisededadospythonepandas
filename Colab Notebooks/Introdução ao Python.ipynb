```python
# Criando variável do tipo string
frase = "Python para análise de dados"
```

```python
# Pegando a primeira letra da frase e a última
frase[0]

```

```python
# Pegando a ultima letra da frase
frase[-1]
```

```python
# Pegando as 10 primeiras letras
frase[0:10]
```

```python
# Pegando as letras da posição 10 para frente
frase[10:]
```

```python
# Pegando as letras ate a posição 17
frase[:17]
```

```python
# Verificando quantas vezes a palavra dados aparece na frase
frase.count('dados')
```

```python
# Deixando todas as letras em minusculo
frase.lower()
```

```python
# Deixando todas as letras em maiusculo
frase.upper()
```

```python
# Trocando alguma palavra da frase
frase.replace('Python', 'Frases')
```
